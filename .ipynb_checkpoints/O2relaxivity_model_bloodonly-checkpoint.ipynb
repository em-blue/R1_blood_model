```python
muirtemp = pd.read_csv('O2relax_temp.csv')
muirtemp


y=muirtemp.r1
temp=muirtemp.Temp

def O2_relaxivity_temp(x,y,temp):

    return x[0] + x[1]*temp -y

x0 = np.array([1.0,1.0])
O2_relax_temp = least_squares(O2_relaxivity_temp, x0, args=(y,temp))
```

```python
temp_slope=O2_relax_temp.x[1]
```

```python
watersaline = pd.read_csv('O2relax_salineonly.csv')
watersaline

nrepetitions=1000
nsamples=math.floor(len(watersaline)*0.7)
ntest=len(watersaline)-nsamples

list_of_br1Ox=np.zeros(nrepetitions)
list_of_mr1Ox_b0=np.zeros(nrepetitions)
list_of_mr1Ox_temp=np.zeros(nrepetitions)
list_of_y_guess_r1=np.zeros((nrepetitions,ntest))
list_of_y_test_r1=np.zeros((nrepetitions,ntest))
list_of_B0_test_r1=np.zeros((nrepetitions,ntest))
list_of_temp_test_r1=np.zeros((nrepetitions,ntest))


B0_alldata_r1=watersaline.B0
y_alldata_r1=watersaline.r1
temp_alldata_r1=watersaline.Temp


for i in range(0,nrepetitions):
    B0,B0_test,y,y_test,temp,temp_test = train_test_split(B0_alldata_r1,y_alldata_r1,temp_alldata_r1, test_size=0.3)

    def O2_relaxivity(x,y,B0,temp):

        return x[0] + x[1]*np.log(B0) + temp_slope*temp -y

    x0 = np.array([1.0,1.0])
    O2_relaxivity_watersaline = least_squares(O2_relaxivity, x0, args=(y,B0,temp))

    y_guess=O2_relaxivity_watersaline.x[0]+O2_relaxivity_watersaline.x[1]*np.log(B0_test)+temp_slope*temp_test 


    list_of_y_test_r1[i]=y_test
    list_of_y_guess_r1[i]=y_guess
    list_of_B0_test_r1[i]=B0_test
    list_of_temp_test_r1[i]=B0_test
    
    list_of_br1Ox[i] =O2_relaxivity_watersaline.x[0]
    list_of_mr1Ox_b0[i] =O2_relaxivity_watersaline.x[1]
```

```python
br1Ox_mean=np.mean(list_of_br1Ox)
mr1Ox_b0_mean=np.mean(list_of_mr1Ox_b0)

br1Ox_std=np.std(list_of_br1Ox)
mr1Ox_b0_std=np.std(list_of_mr1Ox_b0)


print('br1Ox mean =',br1Ox_mean)
print('mr1Ox_b0 mean =',mr1Ox_b0_mean)


print('br1Ox std =',br1Ox_std)
print('mr1Ox_b0 std =',mr1Ox_b0_std)



lowerCI_br1Ox = br1Ox_mean - 1.96*br1Ox_std
upperCI_br1Ox = br1Ox_mean + 1.96*br1Ox_std

lowerCI_mr1Ox_b0 = mr1Ox_b0_mean - 1.96*mr1Ox_b0_std
upperCI_mr1Ox_b0 = mr1Ox_b0_mean + 1.96*mr1Ox_b0_std


print(lowerCI_br1Ox )
print(upperCI_br1Ox)

print(lowerCI_mr1Ox_b0)
print(upperCI_mr1Ox_b0)


all_variables_o_r1=np.array([list_of_br1Ox,list_of_mr1Ox_b0])
all_variables_r1=all_variables_o_r1.T
np.shape(all_variables_r1)

all_variables_df_r1 = pd.DataFrame(all_variables_r1, columns = ['br1Ox','mr1Ox_b0'])

sns.set(font_scale=1.5)
g = sns.catplot(data=all_variables_df_r1,kind='violin',scale='width',palette="crest",inner="quartile",bw=0.2);
fig = matplotlib.pyplot.gcf()
plt.ylabel('Value Estimation Distribution')
plt.xlabel('Parameter Name')
plt.title('Parameter Estimations')
```

```python
from matplotlib.colors import Normalize 
from scipy.interpolate import interpn

def density_scatter( x , y, ax = None, sort = True, bins = 20, **kwargs )   :
    
    if ax is None :
        fig , ax = plt.subplots()
    data , x_e, y_e = np.histogram2d( x, y, bins = bins, density = True )
    z = interpn( ( 0.5*(x_e[1:] + x_e[:-1]) , 0.5*(y_e[1:]+y_e[:-1]) ) , data , np.vstack([x,y]).T , method = "splinef2d", bounds_error = False)

    #To be sure to plot all data
    z[np.where(np.isnan(z))] = 0.0

    # Sort the points by density, so that the densest points are plotted last
    if sort :
        idx = z.argsort()
        x, y, z = x[idx], y[idx], z[idx]

    ax.scatter( x, y, c=z, **kwargs )

    norm = Normalize(vmin = np.min(z), vmax = np.max(z))
    return ax



flat_list_of_y_test_r1=list_of_y_test_r1.flatten()
flat_list_of_y_guess_r1=list_of_y_guess_r1.flatten()
flat_list_of_B0_test_r1=list_of_B0_test_r1.flatten()

y_error_r1=flat_list_of_y_test_r1-flat_list_of_y_guess_r1
mean_r1error=np.mean(y_error_r1)
std_r1error=np.std(y_error_r1)

crude_lower_limit_r1 = mean_r1error-1.96*std_r1error
crude_upper_limit_r1 = mean_r1error+ 1.96*std_r1error


sns.set(font_scale=1)
flat_list_of_y_test_r1=list_of_y_test_r1.flatten()
flat_list_of_y_guess_r1=list_of_y_guess_r1.flatten()
flat_list_of_B0_test_r1=list_of_B0_test_r1.flatten()

y_error_r1=flat_list_of_y_test_r1-flat_list_of_y_guess_r1
mean_r1error=np.mean(y_error_r1)
std_r1error=np.std(y_error_r1)


density_scatter(flat_list_of_y_test_r1, flat_list_of_y_guess_r1, bins = [50,50] )
plt.plot(flat_list_of_y_test_r1,flat_list_of_y_test_r1,'k')
plt.xlabel('r1 Empirical Datapoint')
plt.ylabel('r1 Model Estimate')
```

```python
from sklearn.metrics import mean_squared_error
sns.set(font_scale=1.4)

watersaline = pd.read_csv('O2relax_salineonly.csv')
watersaline

n=2.7
P50=37
Hb=5.15

B0_alldata_r1=watersaline.B0
y_alldata_r1=watersaline.r1
temp_alldata_r1=watersaline.Temp


y_pred=br1Ox_mean+mr1Ox_b0_mean*np.log(B0_alldata_r1)+temp_slope*temp_alldata_r1
y_true=y_alldata_r1

meanSE=mean_squared_error(y_true, y_pred)
print('MSE is', meanSE)

y_pred=np.asarray(y_pred)
y_true=np.asarray(y_true)

y_pred_reshape=y_pred.reshape(-1, 1)
y_true_reshape=y_true.reshape(-1, 1)

reg_v_true_final = LinearRegression().fit(y_true_reshape,y_pred_reshape)

plt.scatter(y_true_reshape,y_pred_reshape,color='darkcyan',s=100)
print('Estimate vs True R2=',reg_v_true_final.score(y_true_reshape,y_pred_reshape))
plt.plot(y_true,y_true,'k')
plt.plot(y_true,np.squeeze(reg_v_true_final.predict(y_true_reshape)),'k',linestyle='dotted',color='darkcyan')
plt.plot(y_true,np.squeeze(reg_v_true_final.predict(y_true_reshape)-1.96*std_r1error),color='darkcyan',alpha=0.2)
plt.plot(y_true,np.squeeze(reg_v_true_final.predict(y_true_reshape)+1.96*std_r1error),color='darkcyan',alpha=0.2)

plt.text(3,1, '$R^2=0.60$')
plt.ylim(0,5)
plt.xlim(1,4)
plt.text(3,0.4, '$MSE=0.18$')
plt.xlabel('r1Ox Empirical Datapoint (1/s)')
plt.ylabel('r1Ox Model Estimate (1/s)')
plt.title('Saline Dataset \n True r1Ox vs Estimation r1Ox')
```
